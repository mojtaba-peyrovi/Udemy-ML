```python
column_names = ['user_id','item_id','rating','timestamp']
```

```python
df = pd.read_table(path + '\\' + 'ratings.dat', names=column_names, delimiter="::")
```

```python
df.head()
```

```python
movie_titles = pd.read_table('movies.dat', names=['item_id','movie_title'], delimiter="::", index_col=False)
```

```python
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles, on='item_id')
```

```python
df.head()
```

```python
list= []
for i in range(1,6):
    count = df[(df.rating == i)].count()    
    list.append(count['movie_title'])  
   
```

```python
movies_by_group = pd.DataFrame(list, index=[1,2,3,4,5],columns=["Ratings"])
movies_by_group
```
